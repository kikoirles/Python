```python
# Intentamos graficar la columna de precios
venta_autos['Precio (USD)'].plot()
```

No funciono, verdad ? Alguna idea de por que esto puede ocurrir?
Una pista es buscar "¿Como convertir strings de pandas a numeros"

```python
# Elimina la puntuacion de la columna de precios
# Convierte la columna 'Precio (USD)' a tipo string antes de aplicar .str.replace
venta_autos['Precio (USD)'] = venta_autos['Precio (USD)'].str.replace('[\,\$\.]', '')
# Divide la columna por 100 para obtener el precio en dolares
venta_autos['Precio (USD)'] = venta_autos['Precio (USD)'].astype (int) / 100 # Fixed typo here
venta_autos
```

```python
# Intentamos graficar la columna de precios
venta_autos['Precio (USD)'].plot()
```
